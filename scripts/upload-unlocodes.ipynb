```python
import pandas as pd
import glob
import os

# load the csv's from UN official publication
path = r'loc212csv'
filenames = glob.glob(os.path.join(path , "2021-2 UNLOCODE CodeListPart1 copy.csv"))
#files = os.path.join('loc212csv', '2021-2 UNLOCODE CodeListPart*.csv')

#files = glob.glob(files)
#print(files)

li = []

for filename in filenames:
    df = pd.read_csv(filename, index_col=None, header=0, encoding='ISO-8859-1')
    li.append(df)
    
locodes_df = pd.concat(li, axis=0, ignore_index=True)

# joining files with concat and read_csv
#df = pd.concat(map(pd.read_csv, files), ignore_index=True)
print(locodes_df)

# @TODO Check Encoding example is `Sant Julia de Loria` shows in the name 
# as `Sant Juli√† de L√≤ria`
```

```python
# The countries are listed in the locodes list, we need to remove these
# they have a value in country and name column, the name always begins
# with . i.e. `.France`

locodes_df = locodes_df.dropna(subset=['code'])
```

```python
# drop unecessary columns that we wont use in a github environment as 
# they are already managed by version control i.e. change and date

locodes_df.drop('change', axis=1, inplace=True)
locodes_df.drop('date', axis=1, inplace=True)

# @TODO look at the output some #NAME? values in function ?
```

```python
# For the records without Decimal Degree co-ordinates
# but with the DMS Coordinates convert to Decimal Degrees
# This should really be included in the maintenance proceedure
# for new codes BUT here we are doing an initial import of UNLOCODE

```

```python
# Create the combined UNLOCODE and replace the contents of the code 
# field similar to UN Website Publishing
# Open question on the space.. systems would remove this it is for humans
print(locodes_df)
locodes_df['locode'] = locodes_df['country'] + locodes_df['code']

```

```python
# Order Outputs to match UN Site 

column_names = ['country', 'locode', 'name', 'name_wo', 'sub_division', 'function', 'status', 'iata', 'coordinates', 'remarks']
locodes_df = locodes_df.reindex(columns=column_names)

print(locodes_df)

```

```python
# split the files by country to create specific files 
# loop through all unique country code values and write to a specific 
# csv file named after the ISO country code

countries = locodes_df.country.unique()
print(countries)

for country in countries:
    country_df = locodes_df.loc[
                (locodes_df['country'] == country)
            ]
    print(country_df)
    country_df.to_csv('locodes/' + country + '.csv', index=False)
```

```python
# Write the complete UNLOCODE dataset into a single file
locodes_df.to_csv('locodes/all-unlocodes.csv', index=False)  
```
